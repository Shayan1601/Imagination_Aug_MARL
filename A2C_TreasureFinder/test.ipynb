```python
from env_shayan import  TreasureFinderEnv
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env

import os
import shutil



log_dir = "./a2c_traning/fixed_log"
if os.path.exists(log_dir):
    shutil.rmtree(log_dir)  # Delete the folder and its contents
os.makedirs(log_dir)  # Create an empty folder





def create_custom_env():
    # Assume bars is defined or passed as a global variable
    env = TreasureFinderEnv(map_size=7)
    return env


vec_env = make_vec_env(create_custom_env, seed=1, n_envs=5)



```

```python
model = A2C("MlpPolicy", vec_env, verbose=1, tensorboard_log=log_dir)
model.learn(total_timesteps=500000)
```
